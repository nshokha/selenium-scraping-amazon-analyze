```python
import pandas as pd
df = pd.read_csv('laptops1.csv')
df.info()
#Higher-priced laptops tend to have higher customer ratings  True
#Premium laptop brands like HP and Dell receive higher average ratings than lesser-known brands. False
#Laptops running Windows 11 have higher ratings than those running older versions (like Windows 10). True
#Laptops with longer customer reviews tend to have higher overall ratings.   False
#Laptops within a mid-range price ($500-$800) receive higher ratings than budget or premium models  True
#Laptops with 16 GB rams tend to have higher overall reviews. True
#Laptops with 512 GB storages tend to have higher price. True
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
high_price = df[df['prices'] >= 200]
low_price = df[df['prices'] < 200]
```

```python
bar_price = high_price['ratings'].value_counts(ascending=False)
print(high_price['ratings'].mean())
```

```python
bar_low_price = low_price['ratings'].value_counts(ascending=False)
print(low_price['ratings'].mean())
```

```python
bar_price.plot(kind='barh')
```

```python
bar_low_price.plot(kind='barh')
```

```python
prem = df[df['brands'].isin(['HP', 'Dell'])]
prem_bar = prem['ratings'].value_counts(ascending=False)
print(prem['ratings'].mean())
```

```python
others = df[~df['brands'].isin(['HP', 'Dell'])]
others_bar = others['ratings'].value_counts(ascending=False)
print(others['ratings'].mean())
```

```python
prem_bar.plot(kind='barh')
```

```python
others_bar.plot(kind='barh')
```
